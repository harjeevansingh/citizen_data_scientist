```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
# 1.Upload the image into python environment and show the image. 

img = imread("phantom.png")
plt.imshow(img)
```

```python
# 2. Perform black and white thresholding with decision boundary set at 0.15. Store the new image in a variable new_img and show it.
# in int we have range [1-255] and in float we have range [0-1]... so in current case we use 1-255 range 

new_img = img.copy()
new_img[new_img>0.15*255] = 255
new_img[new_img<=0.15*255] = 0

plt.imshow(new_img)
```

```python
# 3. Use the numpy array manipulating function fliplr() to get the mirror image.

img_flip = new_img.copy()
img_flip = np.fliplr(img_flip)
plt.imshow(img_flip)
```

```python
# 5. Compress the flipped image along both the axes by 50% in sequence. 

img_cmp = img_flip.copy()
img_cmp = np.compress([True]*200, img_cmp, axis=0)
img_cmp = np.compress([True]*200, img_cmp, axis=1)
plt.imshow(img_cmp)
```
